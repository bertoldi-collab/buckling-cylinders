```python
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.signal import savgol_filter
```

```python
version = 119
fname = f'4fold-imperfection-{version}_eig_freq.txt'
data_all = np.loadtxt(f'data_out/{fname}')
temp_all = data_all[:,0]
freq_all = data_all[:,1:]
```

```python
fig, ax = plt.subplots(figsize = (10,6))

ax.plot(-3*temp_all, freq_all, 'ko')
ax.plot(-3*temp_all, freq_all, 'k-')
ax.set_xlabel('$\Delta V / V_0$', fontsize = 16)
ax.set_ylabel('Eig Freq', fontsize = 16)
```
